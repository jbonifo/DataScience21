```python
#import an read big-mart data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Read data into pandas table - details on pandas in block 3
data = pd.read_csv("../../DATA/company_sales_data.csv") 
```

```python
#access data rows by name
plt.plot(data.toothpaste)
```

## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit

### The result should look something like this:
<img src="IMG/1.png" width=400>

```python
plt.plot(data.month_number,data.total_profit)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title("Company profit per month")
plt.xlabel('Month Number')
plt.ylabel('Total profit')
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


### The result should look something like this:
<img src="IMG/2.png" width=400>

```python
plt.plot(data.month_number,data.total_profit, color="red", lw=3, ls='--', marker='o',markerfacecolor="blue",label="Profit data of last year")
plt.legend(loc=4); # upper left corner
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.xticks(data.month_number)
plt.title("Company sales data of last year")
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).


### The result should look something like this:
<img src="IMG/3.png" width=400>

```python
plt.plot(data.month_number,data.bathingsoap, color="tab:red", lw=3, ls='-', marker='o',markerfacecolor="tab:red",label="bathingsoap")
plt.plot(data.month_number,data.facecream, color="tab:blue", lw=3, ls='-', marker='o',markerfacecolor="tab:blue",label="facecream")
plt.plot(data.month_number,data.facewash, color="orange", lw=3, ls='-', marker='o',markerfacecolor="orange",label="facewash")
plt.plot(data.month_number,data.toothpaste, color="tab:green", lw=3, ls='-', marker='o',markerfacecolor="tab:green",label="toothpaste")
plt.plot(data.shampoo, color="purple", lw=3, ls='-', marker='o',markerfacecolor="purple",label="shampoo")
plt.plot(data.month_number,data.moisturizer, color="tab:brown", lw=3, ls='-', marker='o',markerfacecolor="tab:brown",label="moisturizer")
plt.legend(loc=2) # upper left corner
plt.title("Sales data")
plt.xticks(data.month_number)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.ylim([0, 18000])
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.


### The result should look something like this:
<img src="IMG/4.png" width=400>

```python
plt.plot(data.month_number,data.toothpaste, color="tab:blue", lw=3, ls='', marker='o',markerfacecolor="tab:blue",label="tooth paste Sales data")
plt.legend(loc=2); # upper left corner
plt.grid(True)
plt.grid(linestyle='dashed', linewidth=1)
plt.xticks(data.month_number)
plt.title("Tooth paste Sales data")
plt.xlabel('Month Number')
plt.ylabel('Number of units Solid')
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


### The result should look something like this:
<img src="IMG/5.png" width=400>

```python
plt.bar(data.month_number,data.facecream, align="edge", width=-0.25, alpha=1,label="Face Cream sales data")
plt.bar(data.month_number,data.facewash, align="edge", width=0.25, alpha=1,label="Face Wash sales data")
plt.legend(loc=2); # upper left corner
plt.title("Facewash and facecream sales data")
plt.grid(True)
plt.grid(linestyle='dashed', linewidth=1)
plt.xticks(data.month_number)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 


### The result should look something like this:
<img src="IMG/6.png" width=400>

```python
plt.pie([data['facecream'].sum(),data['facewash'].sum(),data['toothpaste'].sum(),data['bathingsoap'].sum(),data['shampoo'].sum(),data['moisturizer'].sum()],labels=['FaceCream','FaceWash','ToothPaste','Bathing soap','Shampoo','Moisturizer'],autopct='%.1f%%')
plt.legend(loc=4,bbox_to_anchor=(0.8, 0, 0.5, 1))
plt.title("Sales data")
```
